```python
#### EXTREMLY IMPORTANT NOTE: Given the problems with this website, I had to manually adjust the year and the name 
# of the file. A loop would endup in errors that make this task prone to error.


import time
import random
from selenium import webdriver
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import (
    ElementClickInterceptedException, NoSuchElementException
)

def random_sleep(min_sec, max_sec):
    """Generate a random sleep duration."""
    time.sleep(random.uniform(1, 5))


# Define MAX_RETRIES
MAX_RETRIES = 3  

# Open a txt file to save the results
with open('spain2012.txt', 'a') as df:  

    # Initialize the wait
    wait = WebDriverWait(driver, 10)

    # Scroll down to the dropdown each time, as the page might refresh
    driver.execute_script("window.scrollTo(0, 150);")

    # Select the year
    option_year_xpath = f"//*[@id='ddlYear']/option[@value='2012']"
    option_year = wait.until(EC.element_to_be_clickable((By.XPATH, option_year_xpath)))
    option_year.click()
    random_sleep(1, 8)  

    for month_val in range(1, 13):  
        retries = 0
        success = False

        while not success and retries < MAX_RETRIES:
            try:
                # Select the month
                option_month_xpath = f'//*[@id="ddlMonth"]/option[{month_val}]'
                option_month = wait.until(EC.element_to_be_clickable((By.XPATH, option_month_xpath)))
                option_month.click()
                random_sleep(1, 5)

                # Click the search button
                button_search = driver.find_element(By.XPATH, "//*[@id='btnSearch']")
                ActionChains(driver).move_to_element(button_search).click(button_search).perform()
                random_sleep(1, 5)

                page_num = 1
                while True:
                    # Extract and save links of interest
                    for a in driver.find_elements(By.XPATH, '//*[(@id = "columnContentContainer")]//a'):
                        print(a.get_attribute('href'))
                        df.write(str(a.get_attribute('href')) + '\n')
                    
                    # Check for the existence of the next page, handle the last page case
                    try:
                        button_page_xpath = f"#btnPage{page_num}"
                        button_page = driver.find_element(By.CSS_SELECTOR, button_page_xpath)
                        ActionChains(driver).move_to_element(button_page).click(button_page).perform()
                        random_sleep(1, 5)
                        page_num += 1
                    except NoSuchElementException:
                        # No more pages, break the pagination loop
                        break

                success = True  # If we reached this point, the scraping was successful for the current month

            except (ElementClickInterceptedException, NoSuchElementException) as e:
                print(f"An exception occurred: {e}")
                retries += 1
                if retries >= MAX_RETRIES:
                    print(f"Failed to scrape data for month {month_val} after {MAX_RETRIES} attempts. Moving to the next month.")
                    break  

.
```

```python
import os

# List of file names
file_names = ['spain2004.txt', 'spain2005.txt', 'spain2006.txt', 'spain2007.txt', 'spain2008.txt', 'spain2009.txt',
             'spain2010.txt', 'spain2011.txt', 'spain2012.txt', 'spain2013.txt', 'spain2014.txt', 'spain2015.txt',
             'spain2016.txt', 'spain2017.txt', 'spain2018.txt', 'spain2019.txt', 'spain2020.txt', 'spain2021.txt',
             'spain2022.txt', 'spain2023.txt']  # Add all 20 file names here


directory = 'C:\\Users\\Juan\\'  


# Combined list of URLs
all_urls = []

# Read each file and add its contents to the combined list
for file_name in file_names:
    with open(os.path.join(directory, file_name), 'r') as file:
       
        file_contents = file.readlines()
        all_urls.extend(file_contents)

# Remove duplicates
unique_urls = list(set(all_urls))

# Write the unique URLs to a new file
with open(os.path.join(directory, 'combined_urls.txt'), 'w') as file:
    for url in unique_urls:
        file.write(url)

```

```python
# Function to divide a list into smaller chunks
def divide_into_chunks(lst, num_chunks):
    # Calculate the size of each chunk
    chunk_size = len(lst) // num_chunks
    
    # Store the chunks
    chunks = [lst[i * chunk_size:(i + 1) * chunk_size] for i in range(num_chunks)]
    
    # Add the remaining elements to the last chunk if any
    remaining = len(lst) - chunk_size * num_chunks
    for i in range(remaining):
        chunks[i % num_chunks].append(lst[chunk_size * num_chunks + i])
    
    return chunks

# Read all URLs from the file into a list
file_path = 'combined_urls.txt'  
with open(file_path, 'r') as file:
    all_urls = [url.strip() for url in file.readlines()] 
    
# Divide the URLs into 10 nearly equal parts
number_of_lists = 10
lists_of_urls = divide_into_chunks(all_urls, number_of_lists)

# Print the number of URLs in each list
for i, url_list in enumerate(lists_of_urls):
    print(f"List {i + 1} contains {len(url_list)} URLs.")
```

```python
import time
import random
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException


# Function to scrape data from a given URL
def scrape_url(url):
    driver.get(url)
    time.sleep(5)  

    
    try:
        title = driver.find_element(By.XPATH, '//*[@id="h1Title"]').text
    except NoSuchElementException:
        title = None

    try:
        main_text = driver.find_element(By.XPATH, '//*[@id="MainContent"]').text
    except NoSuchElementException:
        main_text = None

    try:
        date = driver.find_element(By.XPATH, '//*[@id="noticia_sitedate"]').text
    except NoSuchElementException:
        date = None

    return title, main_text, date

for idx, url_list in enumerate(lists_of_urls, start=1): 
    print(f"Processing list {idx}...")


    df = pd.DataFrame(columns=['URL', 'Title', 'Main Content', 'Date'])

    for url in url_list:
        print(f"Scraping {url}...")
        try:
            result = scrape_url(url)
            df = df.append({'URL': url, 'Title': result[0], 'Main Content': result[1], 'Date': result[2]}, ignore_index=True)
            time.sleep(random.uniform(19, 49)) 
        except Exception as e:
            print(f"Error scraping {url}: {e}")
            print("Retrying in 5 to 10 minutes...")
            time.sleep(random.uniform(300, 600))  
            continue  
            
    # Save the DataFrame to a CSV
    csv_filename = f'C:/Users/Juan/list{idx}.csv'  
    df.to_csv(csv_filename, index=False, encoding='utf-8')

    print(f"Finished processing list {idx} and saved to {csv_filename}!")

print("Scraping completed!")
driver.quit()  
```
